# K Nearest Neigbors (KNN) and Principle Component Analysis (PCA)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.decomposition import PCA
import numpy as np

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,stratify=y)
scaler = StandardScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
if(prl > 0):
    print("Train:",x_train_scale,"Test:",x_test_scale,sep="\n")
```

```python
if(prl > 0):
    print(x_train_scale.shape,x_test_scale.shape,sep="\n")
```

```python
pca = PCA(n_components=0.999999)
```

```python
x_train_scale_pca = pca.fit_transform(x_train_scale)
if(prl > 1):
    print(x_train_scale_pca)
```

```python
if(prl > 0):
    print("Variance from each component:",pca.explained_variance_ratio_)
    print("Singular Values:",pca.singular_values_)
```

```python
ncomp = []
for i in range(len(pca.explained_variance_ratio_)):
    ncomp.append(i+1)

fig = plt.figure()
ax = fig.add_subplot()
ax.bar(ncomp, pca.explained_variance_ratio_)
ax.set_xlabel('# Components')
ax.set_ylabel('Variance')
plt.show()
```

```python
pca2 = PCA(n_components=2)
x_train_scale_pca2 = pca2.fit_transform(x_train_scale)
loading = pca2.components_.T
```

```python
import pandas as pd

loading_df = pd.DataFrame(
    loading, 
    columns=['PC1', 'PC2'], 
    index=plt_label
)
if(prl > 0):
    print(loading_df)
```

```python
fig = plt.figure()

for feature in loading_df.index:
    plt.arrow(0,0,loading_df.loc[feature,'PC1'],loading_df.loc[feature,'PC2'],head_width=0.03,head_length=0.03,color='red',alpha=0.7)
    plt.text(loading_df.loc[feature,'PC1']*1.15,loading_df.loc[feature,'PC2']*1.15,feature,color='black', ha='center', va='center', fontsize=8)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.xlabel(f"PC1 ({pca2.explained_variance_ratio_[0]*100:.1f}%)", fontsize=12)
plt.ylabel(f"PC2 ({pca2.explained_variance_ratio_[1]*100:.1f}%)", fontsize=12)
plt.title("PCA Loading Plot")
plt.grid(True, alpha=0.3)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
x_new = np.array([[5, 2.9, 1, 0.2]])
prediction_new = knn.predict(x_new)
if(prl > 0):
    print(prediction_new)
```

```python
predictions_test = knn.predict(x_test)
names = iris.target_names
if(prl > 0):
    print("Predictions Text:",names[predictions_test])
    print("Predictions Numerical:",predictions_test)
    print("True Targets Numerical:",y_test)
```

```python
correct = 0
total = len(y_test)
for i in range(len(y_test)):
    if(y_test[i] == predictions_test[i]):
        correct += 1
if(prl > 0):
    print("Accuracy:",correct/total)
```

```python
knum = 112
knn_m = []
for i in range(knum):
    knn_m.append(KNeighborsClassifier(n_neighbors=i+1))
```

```python
predict = []
if(prl > 1):
    print("Actual Targets:",y_test)
for i in range(knum):
    knn_m[i].fit(x_train,y_train)
    predict.append(knn_m[i].predict(x_test))
    if(prl > 1):
        print(f"Prediction #{i+1}:",predict[i])
```

```python
errors = []
total = len(y_test)
for i in range(knum):
    ncorrect = 0
    for j in range(len(y_test)):
        if(y_test[j] == predict[i][j]):
            ncorrect += 1
    errors.append(1 - (ncorrect/total))

if(prl > 1):
    for i in range(knum):
        print(f"Error #{i+1}:",errors[i])
```

```python
K = []
for i in range(knum):
    K.append(i+1)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(K, errors)
ax.set_xlabel('# Neighbors')
ax.set_ylabel('Error')
plt.show()
```

```python
plt.close()
```

# K-Means Clustering (KMC)

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3,n_init='auto')
cluster_labels = kmeans.fit_predict(x)

centroids = kmeans.cluster_centers_
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x[:, 0], x[:, 1], x[:,2], c=cluster_labels, s=50, alpha=0.45, cmap='viridis')
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:,2], c='red', s=200, alpha=0.9, marker='X', label='Centroids')
ax.set_xlabel(plt_label[0]); ax.set_ylabel(plt_label[1]); ax.set_zlabel(plt_label[2])
plt.legend()
plt.show()
```

```python
if(prl > 0):
    print(cluster_labels)
    print(y)

from scipy.optimize import linear_sum_assignment
from sklearn.metrics import confusion_matrix

y_true = np.array(y)
y_pred = np.array(cluster_labels)
n_classes = max(y_true.max(), y_pred.max()) + 1

cm = confusion_matrix(y_pred, y_true, labels=range(n_classes))
row_ind, col_ind = linear_sum_assignment(-cm)

mapping = dict(zip(row_ind, col_ind))
cluster_labels_matched = np.array([mapping[label] for label in y_pred])

if(prl > 1):
    print(mapping)
if(prl > 0):
    print(cluster_labels_matched)

num_correct_arr = y == cluster_labels_matched

if(prl > 1):
    print(num_correct_arr)
```

```python
from sklearn.metrics import adjusted_rand_score
score = adjusted_rand_score(y, cluster_labels)
if(prl > 0):
    print("Cluster Quality (1 is perfect, 0 is random):",score)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x[:, 0], x[:, 1], x[:,2], c=num_correct_arr, s=50, alpha=0.45, cmap='viridis')
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:,2], c='red', s=200, alpha=0.9, marker='X', label='Centroids')
ax.set_xlabel(plt_label[0]); ax.set_ylabel(plt_label[1]); ax.set_zlabel(plt_label[2])
plt.legend()
plt.show()
```

```python
kmeans_ncluster = []
ncluster_labels = []
centroid_ncluster = []
inertia = []
ncluster = 10
Kclust = np.arange(1,ncluster+1)
for i in range(ncluster):
    km = KMeans(n_clusters=i+1, n_init='auto')
    kmeans_ncluster.append(km)
    ncluster_labels.append(km.fit_predict(x))
    centroid_ncluster.append(km.cluster_centers_)
    inertia.append(km.inertia_)

if(prl > 1):
    print(kmeans_ncluster)
    print(ncluster_labels)
if(prl > 0):
    print(inertia)
```

```python
for i in range(ncluster):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')
    ax.scatter(x[:, 0], x[:, 1], x[:,2], c=ncluster_labels[i], s=50, alpha=0.45, cmap='viridis')
    ax.scatter(centroid_ncluster[i][:, 0], centroid_ncluster[i][:, 1], centroid_ncluster[i][:,2], c='red', s=200, alpha=0.9, marker='X', label='Centroids')
    ax.set_xlabel(plt_label[0]); ax.set_ylabel(plt_label[1]); ax.set_zlabel(plt_label[2])
    ax.set_title(f"{i+1} Clusters with Associated Centroids")
    plt.legend()
    plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(Kclust,inertia)
ax.set_xlabel('# Clusters'); ax.set_ylabel('Inertia')
plt.show()
```

```python
plt.close()
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

svm = SVC(kernel='rbf',C=1.0,gamma='scale')
svm.fit(x_train_scale_pca2,y_train)
```

```python
x_test_scale_pca2 = pca2.transform(x_test_scale)
predict_svm = svm.predict(x_test_scale_pca2)
if(prl > 0):
    print(f"Accuracy: {accuracy_score(y_test, predict_svm):.2f}")
    print(classification_report(y_test, predict_svm))
```

```python
from sklearn.inspection import DecisionBoundaryDisplay

fig = plt.figure()

plt.scatter(x_train_scale_pca2[:, 0], x_train_scale_pca2[:, 1], c=y_train, cmap=plt.cm.Paired, edgecolors="k")

DecisionBoundaryDisplay.from_estimator(
    svm,
    x_train_scale_pca2,
    response_method="decision_function",
    plot_method="contour",
    levels=[0],  # Level 0 is the exact separating line
    multiclass_colors="viridis",
    linestyles=["-"],
    ax=plt.gca()
)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
svm_c_arr = []
c_iter = 25
c_vals = []
for i in range(c_iter):
    ci = (i+1)
    c_vals.append(ci)
    svm_c_arr.append(SVC(kernel='rbf',C=ci,gamma='scale'))
if(prl > 1):
    print(svm_c_arr)
```

```python
predict_c_arr = []
accuracy_c = []
for i in range(c_iter):
    svm_c_arr[i].fit(x_train_scale,y_train)
    temp_arr_c = svm_c_arr[i].predict(x_test_scale)
    predict_c_arr.append(temp_arr_c)
    count_c = 0
    for j in range(len(y_test)):
        if(y_test[j] == temp_arr_c[j]):
            count_c += 1
    accuracy_c.append(count_c/len(y_test))
if(prl > 1):
    print(y_test)
    print(predict_c_arr)
    print(accuracy_c)

fig = plt.figure()

plt.plot(c_vals,accuracy_c)
plt.xlabel("C")
plt.ylabel("Accuracy")
```

```python
svm_g_arr = []
g_iter = 6
g_vals = []
for i in range(g_iter):
    gi = 0.0001*(10**(i))
    g_vals.append(gi)
    svm_g_arr.append(SVC(kernel='rbf',C=1.0,gamma=gi))
if(prl > 1):
    print(svm_g_arr)
```

```python
predict_g_arr = []
accuracy_g = []
for i in range(g_iter):
    svm_g_arr[i].fit(x_train_scale,y_train)
    temp_arr_g = svm_g_arr[i].predict(x_test_scale)
    predict_g_arr.append(temp_arr_g)
    count_g = 0
    for j in range(len(y_test)):
        if(y_test[j] == temp_arr_g[j]):
            count_g += 1
    accuracy_g.append(count_g/len(y_test))
if(prl > 1):
    print(y_test)
    print(predict_g_arr)
    print(accuracy_g)

fig = plt.figure()

plt.plot(g_vals,accuracy_g)
plt.xscale('log')
plt.xlabel("log(gamma)")
plt.ylabel("Accuracy")
```

```python
svm_gc_arr = []
for i in range(g_iter):
    for j in range(c_iter):
        svm_gc_arr.append(SVC(kernel='rbf',C=c_vals[j],gamma=g_vals[i]))
if(prl > 1):
    print(svm_gc_arr)
```

```python
predict_gc_arr = []
accuracy_gc = []
for i in range(g_iter):
    for j in range(c_iter):
        svm_gc_arr[i*c_iter+j].fit(x_train_scale,y_train)
        temp_arr_gc = svm_gc_arr[i*c_iter+j].predict(x_test_scale)
        predict_gc_arr.append(temp_arr_gc)
        count_gc = 0
        for k in range(len(y_test)):
            if(y_test[k] == temp_arr_gc[k]):
                count_gc += 1
        accuracy_gc.append(count_gc/len(y_test))
if(prl > 1):
    print(y_test)
    print(predict_gc_arr)
    print(accuracy_gc)
```

```python
accuracy_gc = np.array(accuracy_gc,dtype=float)
accuracy_gc = accuracy_gc.reshape(g_iter,c_iter)
if (prl > 1):
    print(accuracy_gc)
C, G = np.meshgrid(c_vals, np.log(g_vals))  

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot_surface(C,G,accuracy_gc)
ax.set_xlabel("C")
ax.set_ylabel("log(gamma)")
ax.set_zlabel("Accuracy")
```
